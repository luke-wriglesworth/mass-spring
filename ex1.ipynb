```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16, 'figure.figsize': (12, 9), 'font.family': 'serif'})

dt = 0.01
k = 2
m = 5
omega_0 = np.sqrt(k/m)
def simulate(y_0, v_0, omega_0, dt, t_max):
    t = np.linspace(0,t_max, int(t_max/dt))
    y = np.zeros(len(t))
    v = np.zeros(len(t))  
    y[0] = y_0
    v[0] = v_0
    for i in range(1,len(t)):
        y[i] = y[i-1] + dt*v[i-1]
        v[i] = v[i-1] - dt*omega_0**2*y[i-1]
    return y, v, t

y, v, t = simulate(5, 2, omega_0, dt, 100)
plt.title("Forward Euler Simulation")
plt.scatter(t,y, label="y(t)")
plt.scatter(t,v, label="v(t)")
plt.xlabel("Time [s]")
plt.ylabel("Distance")
plt.legend()
plt.grid()
plt.show()
```
